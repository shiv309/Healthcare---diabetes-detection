# Validation with train_test_split, KFold and StratifiedKFold

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_test,y_test)

lr_pred = lr.decision_function(X_test)
```

```python
svm = SVC()
svm.fit(X_train, y_train)
svm.score(X_test,y_test)

svm_pred = svm.decision_function(X_test)
```

```python
RF = RandomForestClassifier(n_estimators= 50, random_state= 123)
RF.fit(X_train, y_train)
RF.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
score_lr = cross_val_score(lr, X, y, cv = 5)
score_RF = cross_val_score(RF, X, y, cv = 5)
score_svm = cross_val_score(svm, X, y, cv = 5)
```

```python
print(score_lr)
score_lr.mean()
```

```python
print(score_RF)
score_RF.mean()
```

```python
print(score_svm)
score_svm.mean()
```

```python
from sklearn.model_selection import StratifiedKFold
Skf = StratifiedKFold(n_splits = 5, random_state= 890)
Skf
```

```python
for train_index, test_index in Skf.split(X,y):
    print("Train:", train_index, "Validation:", test_index)
    X1_train, X1_test = X.iloc[train_index], X.iloc[test_index]
    y1_train, y1_test = y.iloc[train_index], y.iloc[test_index]
```

```python
def get_score(model, X1_train, y1_train):
    model.fit(X1_train, y1_train)
    print(model.score(X1_test, y1_test))
```

```python
print('Score of LinearRegression:'),get_score(lr, X1_train, y1_train)
print('Score of SVM:'),get_score(svm, X1_train, y1_train)
print('Score of RandomForest:'),get_score(RF, X1_train, y1_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 10)
```

```python
classifier.fit(X_train, y_train)
pred = classifier.predict(X_test)
```

```python
classifier.fit(X1_train, y1_train)
pred1 = classifier.predict(X1_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y1_test, pred1))
print(classification_report(y1_test, pred1))
```

```python
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```


 # ROC and AUC curve plot

```python
from sklearn.metrics import roc_curve, auc
lr_fpr, lr_tpr, threshold = roc_curve(y_test, lr_pred)
auc_lr = auc(lr_fpr, lr_tpr)
auc_lr
```

```python
svm_fpr, svm_tpr, threshold = roc_curve(y_test, svm_pred)
auc_svm = auc(svm_fpr, svm_tpr)
auc_svm
```

```python
plt.figure(figsize= (5,5), dpi = 100)
plt.plot(lr_fpr, lr_tpr, linestyle = '-', label = 'Logistic (auc = %0.3f)' % auc_lr)
plt.plot(svm_fpr, svm_tpr, marker = '.', label = 'SVM (auc = %0.3f)' % auc_svm)

plt.legend()
```
